**Option Pricing using the Black-Scholes-Merton Model**

In finance, The Black-Scholes-merton model is one of the most widely used methods for pricing options. It calculates the theoretical value of an option based on five key variables



*   Underlying Price (S)
*   Strike Price (K)
*   Time to Expiration (T)
*   Risk Free Rate (r)
*   Volatility








**Import necessary libraries**

```python
import math
from scipy.stats import norm
```

Define the variables

```python
# Define the Variables
S = 42 #Underlying Price
K = 40 #strike Price
T = 0.5 #Time to Expiration
r = 0.1 #Risk-Free Rate
Vol= 0.2 #Volatility
```

Screenshot 2025-08-14 191111.png

```python
d1 = (math.log(S/K) + (r + 0.5 * Vol**2)*T) / (Vol * math.sqrt(T))
```

Screenshot 2025-08-14 193012.png

```python
d2 = d1 - (Vol * math.sqrt(T))
```

Screenshot 2025-08-14 193323.png

```python
C = S * norm.cdf(d1) - K * math.exp(-r*T)* norm.cdf(d2)
```

Screenshot 2025-08-14 193549.png

```python
P = K * math.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
```

Print the results

```python
print('The value of d1 is: ', round(d1, 4))
print('The value of d2 is: ', round(d2, 4))
print('The price of the call option is: $', round(C, 2))
print('The price of the put option is: $', round(P, 2))
```
